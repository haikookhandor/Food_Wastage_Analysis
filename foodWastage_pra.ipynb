```python
import pandas as pd

# read "mess_wastage_2020.csv"
mess_wastage = pd.read_csv('mess_wastage_2019.csv')

mess_wastage.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Convert the 'Date' column to a datetime object
mess_wastage['Date'] = pd.to_datetime(mess_wastage['Date'], format='%m-%d-%y')

# Set the x-axis to display the months
fig, ax = plt.subplots(figsize=(20,10))
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))

# Plot the data
ax.plot(mess_wastage['Date'], mess_wastage['Breakfast'], color='red')
ax.set_title('Breakfast wastage vs dates of 2019', fontsize=14)
ax.set_xlabel('Months', fontsize=14)
ax.set_ylabel('Breakfast wastage (in KG)', fontsize=14)
ax.grid(True)

# Add vertical lines at the end of each month
# for month in range(1, 13):
#     ax.axvline(mess_wastage[mess_wastage['Date'].dt.month == month]['Date'].max(), color='black')

plt.show()
```
